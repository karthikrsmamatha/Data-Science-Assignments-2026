```python
drop_df=df.drop_duplicates()
drop_df
```

```python
mean_bill=df['BillAmount'].mean()
mean_bill
```

```python
filled_df=drop_df.fillna({'BillAmount': mean_bill})
filled_df
```

```python
merged_df = pd.merge(filled_df, df1, on='PatientID')
merged_df
```

```python
new_patients_data = {
    'PatientID': [106, 107],
    'Name': ['Frank', 'Grace'],
    'Department': ['Neurology', 'Orthopedics'],
    'Doctor': ['Dr. John', 'Dr. Lee'],
    'BillAmount': [4200.0, 5100.0],
    'ReceptionistID': [3, 1],
    'CheckInTime': ['2023-01-15 14:00', '2023-01-16 09:15']
}

new_df = pd.DataFrame(new_patients_data)
new_df
```

```python
con_df=pd.concat([filled_df,new_df],ignore_index=True)
con_df
```

```python
new_billing = pd.DataFrame({
    'DoctorFee': [1500, 1800, 2000, 1500, 1200, 1500, 1800],
    'MedicineCharges': [450, 600, 750, 300, 500, 450, 800]
})

final_df = pd.concat([con_df, new_billing], axis=1)
final_df
```
