#### Scaling

```python
### Select Numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
num_cols
```

```python
std_sca=StandardScaler()
min_max=MinMaxScaler()
```

```python
###Standard Scaler
df_standard = df.copy()
df_standard[num_cols] = std_sca.fit_transform(df_standard[num_cols])
df_standard.head()
```

##### Standard Scaling

**Preferred when:**

* Numerical features have different ranges.
* The data is approximately normally distributed.
* Models such as Logistic Regression, SVM, KNN, PCA, and Linear Regression are used.

**why:**

* Standard Scaling changes the data so that it has a mean of 0 and a standard deviation of 1.
* It ensures that all numerical features are on a similar scale.
* This prevents features with larger values from having a greater influence on the model.
* As a result, many machine learning algorithms can learn more effectively and produce better results.

```python
### min-max Scaler
df_minmax = df.copy()
df_minmax[num_cols] = min_max.fit_transform(df_minmax[num_cols])
df_minmax.head()
```

##### Min-Max Scaling

**Preferred when:**

* Numerical features need to be in the range of 0 to 1.
* Neural Networks and other algorithms that work better with normalized data are used.
* The dataset has very few extreme outliers.

**Why:**

* Min-Max Scaling rescales all numerical features to a fixed range between 0 and 1.
* It keeps the relative differences between data points while making all features have the same scale.
* This helps models that require bounded input values train more efficiently.
* It is a good choice when the original data does not contain many extreme values.

#### 2. Encoding Techniques:

```python
cat_col = [col for col in df.select_dtypes(include='object').columns if df[col].nunique() < 5] ### less than 5 categories
cat_col
```

```python
### One-Hot Encoder
one_hot=OneHotEncoder()
one_hot = pd.get_dummies(df, dtype='int',columns=cat_col, drop_first=True)
one_hot.head()
```

    One-Hot Encoding was applied to categorical variables with less than 5 categories. Each category was converted into a separate binary column with values of 0 and 1. This helps the model treat each category independently without assuming any relationship between them.

```python
### Label Encoding
cat_col = [col for col in df.select_dtypes(include='object').columns if df[col].nunique() >5] ### Greater than 5 categories
cat_col
```

```python
df_label = df.copy()
lab_en = LabelEncoder()
for col in cat_col:
    df_label[col] = lab_en.fit_transform(df_label[col])
    
df_label.head()
```

    Label Encoding was applied to categorical variables with more than 5 categories. Each unique category was converted into a unique numerical value. This reduces the number of columns compared to One-Hot Encoding and makes the dataset more memory-efficient.

#### Pros and Cons of One-Hot Encoding
**Pros**
* Easy to understand and implement.
* Does not create any order between categories.
* Works well for categorical variables with a small number of categories.
* Suitable for many machine learning algorithms.
  
**Cons**
* Creates additional columns for each category.
* Increases the size of the dataset when there are many categories.
* Can increase memory usage and computation time.

#### Pros and Cons of Label Encoding

**Pros**
* Easy to apply.
* Does not increase the number of columns.
* Uses less memory than One-Hot Encoding.
* Suitable for categorical variables with many categories.
  
**Cons**
* Assigns numerical values to categories, which may create an artificial order.
* Some machine learning algorithms may incorrectly assume that larger numbers represent higher importance.
* May reduce model performance for nominal categorical variables.

#### 3. Feature Engineering:

```python
### Age Group
df_feature = df.copy()

df_feature['Age_Group'] = pd.cut(
    df_feature['age'],
    bins=[0, 25, 40, 60, 100],
    labels=['Young', 'Adult', 'Middle_Aged', 'Senior']
)

df_feature.head()
```

    Age_Group was created by grouping individuals based on their age. People in different age groups may have different income levels and work experience, making this feature useful for prediction.

```python
### Total capital
df_feature['Total_Capital'] = df_feature['capital_gain'] - df_feature['capital_loss']

df_feature.head()
```

    Total_Capital was created by combining capital_gain and capital_loss. It represents the overall capital value of an individual, which may have a stronger relationship with income than considering these two features separately.

```python
num_cols = df_feature.select_dtypes(include=['int64', 'float64']).columns

df_feature[num_cols].skew()
```

#### Log Transformation
    The capital_gain feature was selected for log transformation because it is highly positively skewed (skewness = 11.95). Most observations have a value of 0, while a few observations have very large values.

    Applying log1p() reduces the skewness, decreases the impact of extreme values, and creates a more balanced distribution. This can improve the performance of machine learning models.

```python
df_feature['capital_gain_log'] = np.log1p(df_feature['capital_gain'])
df_feature.head()
```
