# Hypothesis Testing

### Objective:
To investigate the restaurant owner's claim about the increase in weekly operating costs using hypothesis testing.


```python
import numpy as np
import scipy.stats as stats
import statsmodels
import warnings
warnings.filterwarnings('ignore')
```

#### 1. Hypothesis Statements

*   **Null Hypothesis(H0):** Pop_mean is 4000
*   **Alternative Hypothesis(Ha):** Pop_mean is greater than 4000


```python
### Given Data
pop_mean = 1000 + (5 * 600)=400
pop_std_dev = 5 * 25 = 125     
n = 25                       
x_bar = 3050        
```

#### 2. Test Statistic -- Z score

```python
standard_error = pop_std_dev / np.sqrt(n)
z_score = (x_bar - pop_mean) / standard_error
```

```python
z_score
```

#### 3. Probability

```python
alpha=0.05 # 5% significance level 
z_critical = stats.norm.ppf(1 - alpha)  
z_critical
```

```python
z_critical < z_score
```

#### 4. Decision 
z_critical < z_score is Fasle
#### Accept the Null Hypothesis (H0)

#### 5. Conclusion

There is no evidence to support the restaurant owners' claim. The data shows that weekly operating costs are not higher than what the model predicts.

* **Why?** The old model predicts a cost of **Rs. 4,000**. The owners' actual average cost is only **Rs. 3,050**. 
* **Result:** Because their real costs (3,050) are actually lower than the model's estimate (4,000), the owners' complaint is incorrect. The current cost model is still perfectly fine to use.
