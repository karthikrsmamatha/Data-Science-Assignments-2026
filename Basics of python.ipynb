## Exercise 1: Prime Numbers

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:  
        return True
    if n % 2 == 0: 
        return False

    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True


num = int(input("Enter a number to check Prime or not: "))
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

## Exercise 2: Product of Random Numbers

```python
import random
num1=random.randint(1,50)
num2=random.randint(1,50)
print(f'The two random numbers generated are {num1} and {num2}')

mult=num1*num2

user_prd=int(input(f'Enter the product of random numbers generated:'))
if user_prd==mult:
    print('The product you entered is correct')
else:
    print(f'The product you entered is incorrect')
    print(f"The  correct product of {num1} and {num2} is {mult}")
```

## Exercise 3: Squares of Even/Odd Numbers

## Exercise 4: Hospital Billing System

```python
Room_chg=2000
lab_fee=300
doct_fee=1500
no_days=int(input("Enter number of days the patient stayed:"))
no_lab_lest=int(input("Enter number of lab test performed:"))
med_chg=int(input("Enter Total medicine charge"))

total_amt= (Room_chg * no_days) + doct_fee + (lab_fee * no_lab_lest) + med_chg
if total_amt<=10000:
    print(f'The final bill to be paid =  {total_amt}')
else:
    print(f'The total amount before discount = {total_amt}')
    dis=total_amt*0.1
    total_amt-=dis
    print(f'The Discount applied is 10% = {dis}')
    print(f'The final bill to be paid = {total_amt}')
```
